```python
df_numero_clientes_con_producto["ant"] = df_numero_clientes_con_producto["pk_cid"].shift(-1)
df_numero_clientes_con_producto["dif"] = df_numero_clientes_con_producto["pk_cid"] - df_numero_clientes_con_producto["ant"]
```

<a id = "metrica4"></a>
#  Evolución Conversión
[Volver al índice](#toc)

```python
df_numero_clientes_con_producto
```

En la tabla superior observamos el crecimiento partición a partición, siendo la columna diferencia el diferencial absoluto entre la partición actual y la inmediatamente anterior. Se observa una ralentización en la capacidad de la empresa en aumentar el número de productos por cliente. Con picos de crecimiento destacables en las particiones de julio a noviembre, especialmente en Octubre de 2018. Las dos últimas particiones obtenemos el ritmo de crecimiento más bajo desde que hay registros.

```python
df_numero_clientes = df_clientes.groupby(by=["pk_partition"])[["pk_cid"]].count().sort_values(by="pk_partition", ascending=False)
```

```python
df_evolucion_clientes=pd.concat([df_numero_clientes, df_numero_clientes_con_producto], axis=1)
```

```python
df_evolucion_clientes.head()
```

```python
df_evolucion_clientes.columns = ["total", "con_producto", "siguiente", "dif"]
```

```python
df_evolucion_clientes["porcentaje"] = round(df_evolucion_clientes["con_producto"] / df_evolucion_clientes["total"] * 100, 2)
```

```python
df_evolucion_clientes.reset_index(inplace=True)
```

```python
df_evolucion_clientes
```

Obtenemos el porcentaje de fidelización de nuestros clientes. cuá es el porcentaje partición a partición de clientes con producto. Nuestra conversión ha caído desde el inicio. Los meses de mayor concentración de venta también han sido los meses en los que se han captado en Base de datos nuevos registros, pero no todos han convertido en venta. De ahi la caída de conversión del 98% incial al 74,85% actual. Encontramos el mayor descenso en la tasa de conversión en julio de 2018. Lo vemos de manera gráfica.

```python
fig = px.line(df_evolucion_clientes, x="pk_partition", y="porcentaje", range_y=[50,100])
fig.show()
```

La entrada de julio 2018 de clientes en DB no fué efectiva en términos de conversión, bajando hasta el 76%.
La evolución posterior a julio es similar a la anterior, la entrada en DB parece convertir al ritmo anterior puesto que el 76 baja a 74 progresivamente, pero de forma escalonada y progresiva. 
Es posible que en las dos últimas particiones concentremos el mayo número de bajas.

# Activos Web
[Volver al índice](#toc)

```python
df_numero_clientes_activos = df_clientes[df_clientes["active_customer"] == 1].groupby(by=["pk_partition"])[["pk_cid"]].count().sort_values(by="pk_partition", ascending=False)
```

```python
df_evolucion_clientes_activos_web = pd.concat([df_numero_clientes, df_numero_clientes_activos], axis=1)
df_evolucion_clientes_activos_web.columns = ["total", "activos_web"]
```

```python
df_evolucion_clientes_activos_web
```

```python
df_evolucion_clientes_activos_web["porcentaje"] = round(df_evolucion_clientes_activos_web["activos_web"] / df_evolucion_clientes_activos_web["total"] * 100, 2)
```

```python
df_evolucion_clientes_activos_web.reset_index(inplace=True)
```

```python
df_evolucion_clientes_activos_web
```

```python
fig = px.line(df_evolucion_clientes_activos_web, x="pk_partition", y="porcentaje", range_y=[0,50],title='Evolución de navegación en la web')
fig.show()
```

Inflexión similar en cuanto a clientes activos (navegan en la web en los últimos 3 meses) en julio encontramos una caída del -10%. El comportamiento en cuanto al volumen % de clientes activos en web es similar al % de conversión, con una caída acusada en julio y normalizada a lo largo de los meses siguientes.


```python
anterior_evo_clientes = df_evolucion_clientes["total"].shift(-1)
df_evolucion_clientes["incremento_porcentual"] = (df_evolucion_clientes["total"] - anterior_evo_clientes) / anterior_evo_clientes * 100
anterior_evo_clientes_con_producto = df_evolucion_clientes["con_producto"].shift(-1)
df_evolucion_clientes["incremento_porcentual_con_producto"] = (df_evolucion_clientes["con_producto"] - anterior_evo_clientes_con_producto) / anterior_evo_clientes_con_producto * 100
df_evolucion_clientes
```

En crecimiento porcentuales de clientes en base de datos con y sin productos vemos esa diferenciación. Con un pico de crecimiento de pk_cid del 34% en julio de 2018, frente a un crecimiento del 4% de clientes con producto.
En los meses inmediatos observamos un crecimiento superior en clientes con producto, muy ligado al incremeto porcentual de nuevos pk_cid en base de datos.

```python
df_evolucion_clientes_T = df_evolucion_clientes.melt(id_vars="pk_partition", value_vars=["incremento_porcentual", "incremento_porcentual_con_producto"]).fillna(0)
```

```python
df_evolucion_clientes_T
```

```python
df_evolucion_clientes_T.to_pickle(path='gdrive/MyDrive/EASY_MONEY_NUCLIO/data/evolucion_clientes.pkl')
```

#Clientes en Hibernación: 
Clientes que desde que han entrado en Base de datos no tienen ningún producto contratado. Nunca han tenido un 1.

[Volver al índice](#toc)



```python
df_target = df_total.copy(deep=True)
```

```python
#numero de clientes totales
len(df_target["pk_cid"].unique())
```

```python
# HAY QUE TENER EN CUENTA SI ESTOS CLIENTES POSTERIORMENTE PASAN A TENER UN PRODUCTO
cid_sin_productos = df_target.loc[df_target[products].sum(axis=1) == 0, "pk_cid"].unique()
cid_con_productos = df_target.loc[df_target[products].sum(axis=1) >= 1, "pk_cid"].unique()
```

```python
print(len(cid_sin_productos), len(cid_con_productos))
```

```python
set_cid_sin_productos = set(cid_sin_productos)
interseccion = set_cid_sin_productos.intersection(cid_con_productos)
print(len(interseccion))
```

```python
set_cid_sin_productos_cleaned = set_cid_sin_productos - interseccion
print(len(set_cid_sin_productos_cleaned))
```

```python
df_target_cleaned = df_target[df_target["pk_cid"].isin(set_cid_sin_productos_cleaned)]
```

```python
numero_de_clientes_unicos_sin_producto = len(df_target_cleaned["pk_cid"].unique())
numero_de_clientes_unicos_con_producto = len(set(cid_con_productos))
print(numero_de_clientes_unicos_sin_producto)
print(numero_de_clientes_unicos_con_producto)
```
